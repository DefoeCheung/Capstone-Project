```python
import pandas as pd
import numpy as np
```

```python
# import dataset and check missing values
df = pd.read_csv('StudentsPerformance.csv')
df = df.dropna()
df.columns = ['gender',
              'race_ethnicity',
              'parental_level_of_education',
              'lunch',
              'test_preparation_course',
              'math_score',
              'reading_score',
              'writing_score'
             ]
df
```

```python
# describe the 3 types of score
df.describe()
```

```python
# visualize the outliers of score
df.boxplot()
```

```python
# Remove the score outliers from df
df = df[np.abs(df['math_score']-df['math_score'].mean()) <= (3*df['math_score'].std())]
df = df[np.abs(df['reading_score']-df['reading_score'].mean()) <= (3*df['reading_score'].std())]
df_cleaned = df[np.abs(df['writing_score']-df['writing_score'].mean()) <= (3*df['writing_score'].std())]
df_cleaned
```

```python
# quantify parental_level_of_education
df_cleaned['parental_level_of_education'] = df_cleaned['parental_level_of_education'].map({'some high school': 0,
                                                                                          'high school': 1,
                                                                                          "associate's degree": 2,
                                                                                          'some college': 3,
                                                                                          "bachelor's degree": 4,
                                                                                          "master's degree": 5})
```

```python
# quantify race_ethnicity
dummies1 = pd.get_dummies(df_cleaned.race_ethnicity)
dummies1
```

```python
# quantify gender
dummies2 = pd.get_dummies(df_cleaned['gender'])
dummies2
```

```python
#quantify test_preparation_course
df_cleaned['test_preparation_course'] = df_cleaned['test_preparation_course'].map({'none': 0,'completed': 1})
```

```python
# merge dummies into df
df_quant = pd.concat([df_cleaned,dummies1,dummies2],axis='columns')
df_quant
```

```python
df_quant = df_quant.drop(['gender','race_ethnicity','group A','male'],axis='columns')

df_quant
```

```python
import warnings
warnings.filterwarnings('ignore')
from pymatch.Matcher import Matcher
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
# creat test and control groups
test = df_cleaned[df_quant.lunch == "free/reduced"]
control = df_cleaned[df_quant.lunch == "standard"]
test['lunch'] = 1
control['lunch'] = 0
```

```python
# Matcher
m = Matcher(test, control, yvar="lunch", exclude=['math_score','reading_score','writing_score'])
```

```python
# for reproducibility
np.random.seed(20170925)
m.fit_scores(balance=True, nmodels=350)
```

```python
# pridict scores
m.predict_scores()
m.plot_scores()
```

```python
# tune threshold
m.tune_threshold(method='random')
```

```python
# match data
m.match(method="min", nmatches=1, threshold=0.0004)
m.record_frequency()
```

```python
m.assign_weight_vector()
pairs = m.matched_data.sort_values("match_id")
pairs['total_score'] = pairs['math_score'] + pairs['reading_score'] + pairs['writing_score']
lunch = pairs.loc[pairs['lunch'] == 0]
no_lunch = pairs.loc[pairs['lunch'] == 1]
no_lunch
```

```python
# Calculate mean and std of total_scores of with_lunch group and without_lunch group 
miu1 = np.mean(lunch['total_score'])
miu2 = np.mean(no_lunch['total_score'])
sigma1 = np.std(lunch['total_score'])
sigma2 = np.std(no_lunch['total_score'])
sigma1_2 = np.sqrt((sigma1**2)/350 + (sigma2**2)/350)
print(sigma1_2, miu1-miu2)
```
